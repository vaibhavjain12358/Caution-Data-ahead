```python
# Importing the numpy and pandas packages

import numpy as np
import pandas as pd
```

```python
# Reading csv file

movies = pd.read_csv("../Downloads/Movie+Assignment+Data.csv")
movies
```

```python
# Checking Data Frame
movies.columns
movies.shape
movies.info()
```

```python
# column-wise null count here
movies.isnull().sum()
```

```python
# row-wise null count here
movies.isnull().sum(axis=1)
```

```python
# column-wise null percentages here
print(round(100*(movies.isnull().sum()/len(movies.index)), 2))
```

```python
# dropping unnecessary columns here. 
movies = movies.drop(['color',
'director_facebook_likes',
'actor_1_facebook_likes',
'actor_2_facebook_likes',
'actor_3_facebook_likes',
'actor_2_name',
'cast_total_facebook_likes',
'actor_3_name',
'duration',
'facenumber_in_poster',
'content_rating',
'country',
'movie_imdb_link',
'aspect_ratio',
'plot_keywords'], axis=1)
print(round(100*(movies.isnull().sum()/len(movies.index)), 2))
```

```python
# dropping the rows here
movies = movies[~np.isnan(movies['gross'])]
movies = movies[~np.isnan(movies['budget'])]
round(100*(movies.isnull().sum()/len(movies.index)), 2)
```

```python
# dropping the rows here
movies = movies[movies.isnull().sum(axis = 1) <= 5]
round(100*(movies.isnull().sum()/len(movies.index)), 2)
```

```python
movies.loc[:,'language'].describe()
```

```python
# filling the NaN values in the 'language' column here
movies.loc[pd.isnull(movies['language']), 'language'] = 'English'
round(100*(movies.isnull().sum()/len(movies.index)), 2)
```

```python
# checking number of retained rows here
len(movies.index)/5043
```

```python
# code for unit conversion here
def dollars_to_million(x):
    return x/1000000
movies['budget'] = movies['budget'].apply(dollars_to_million)
movies['gross'] = movies['gross'].apply(dollars_to_million)
movies.loc[:,['gross','budget']]
```

```python
# code for creating the profit column here
movies['profit'] = movies['gross'] - movies['budget']
movies.loc[:,'profit']
```

```python
# sorting the dataframe here
movies.sort_values(by='profit', ascending=False)
```

```python
top10 = top10 = movies.sort_values(by='profit', ascending=False).head(n=10)
top10
```

```python
# dropping duplicate values here
movies.drop_duplicates(inplace=True)
movies
```

```python

top10 = movies.sort_values(by='profit', ascending=False).head(n=10)
top10
```

```python
# code for extracting the top 250 movies as per the IMDb score here. storing it in a new dataframe 
# and name that dataframe as 'IMDb_Top_250'
IMDb_Top_250 = movies.loc[movies.num_voted_users>25000].sort_values(by='imdb_score', ascending=False).head(n=250)
IMDb_Top_250['Rank'] = range(1,251)
IMDb_Top_250
```

```python
Top_Foreign_Lang_Film = IMDb_Top_250.loc[IMDb_Top_250['language'] != 'English']
Top_Foreign_Lang_Film
```

```python
# code for extracting the top 10 directors here
group_by_director = movies.groupby('director_name')
top10director = group_by_director.imdb_score.mean().sort_values(ascending= False).head(n=10)
top10director
```

```python
# code for extracting the first two genres of each movie here
movies['genre_1'] = movies['genres'].str.split('|').str[0]
movies['genre_2'] = movies['genres'].str.split('|').str[1]
movies.loc[:,['genre_1','genre_2']]
movies['genre_2'].fillna(value = movies['genre_1'], inplace=True)
movies.loc[:,['genre_1','genre_2']]
```

```python
movies_by_segment = movies.groupby(['genre_1','genre_2'])
```

```python
PopGenre = movies_by_segment['gross'].mean().sort_values(ascending=False).head(n=5)
PopGenre
```

```python
# code for creating three new dataframes here

Meryl_Streep = movies.loc[movies['actor_1_name'] == 'Meryl Streep']
```

```python
Leo_Caprio = movies.loc[movies['actor_1_name'] == 'Leonardo DiCaprio']
```

```python
Brad_Pitt = movies.loc[movies['actor_1_name'] == 'Brad Pitt']
```

```python
# code for combining the three dataframes here
Combined = Meryl_Streep.append(Leo_Caprio).append(Brad_Pitt)
Combined
```

```python
# code for grouping the combined dataframe here
group_by_actor = Combined.groupby(['actor_1_name'])
```

```python
# code for finding the mean of critic reviews and audience reviews here
group_by_actor['num_critic_for_reviews', 'num_user_for_reviews'].mean().sort_values(by=['num_critic_for_reviews','num_user_for_reviews'] ,ascending=False)
```
